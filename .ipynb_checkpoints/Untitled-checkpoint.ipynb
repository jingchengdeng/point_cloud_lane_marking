```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

```python
def read_data():
    file = open("final_project_point_cloud.fuse", "r")
    data = [f.split() for f in file]
    file.close()
    for d in data:
        for i in range(4):
            d[i] = float(d[i])
    return data
```

```python
data = read_data()
data
```

```python
x, y, z, intensity = np.array(data).T
fig = plt.figure()
plt.hist(z)
plt.show()
```

```python
z_range = max(z) - min(z)
median = np.median(z)
min_z = median - 0.05 * z_range
max_z = median + 0.05 * z_range
horizon_data = []
for d in data:
    if d[2] > min_z and d[2] < max_z:
        horizon_data.append(d)
horizon_x, horizon_y, horizon_z, horizon_intensity = np.array(horizon_data).T
plt.hist(horizon_intensity)
plt.show()
```

```python
plt.plot(horizon_x, horizon_y, ',')
plt.show()
```

```python
lane_mark = []
for d in horizon_data:
    if d[3] >= 70 and d[3] <= 79:
        lane_mark.append(d[:-1])
lane_x, lane_y, lane_z = np.array(lane_mark).T
plt.plot(horizon_x, horizon_y, ',')
plt.plot(lane_x, lane_y, ',', color = "red")
plt.show()
```

```python
kmeans = KMeans(max_iter = 10000, n_clusters = 3)
kmeans.fit(lane_mark)
labels = kmeans.labels_
```

```python
lane_0, lane_1, lane_2 = [], [], []
for i in range(len(labels)):
    if labels[i] == 0:
        lane_0.append(lane_mark[i])
    elif labels[i] == 1:
        lane_1.append(lane_mark[i])
    else:
        lane_2.append(lane_mark[i])
lane_0_x, lane_0_y, lane_0_z = np.array(lane_0).T
lane_1_x, lane_1_y, lane_1_z = np.array(lane_1).T
lane_2_x, lane_2_y, lane_2_z = np.array(lane_2).T
plt.plot(horizon_x, horizon_y, ',', color = "yellow")
plt.plot(lane_0_x, lane_0_y, ',', color = "red")
plt.plot(lane_1_x, lane_1_y, ',', color = "green")
plt.plot(lane_2_x, lane_2_y, ',', color = "blue")
plt.show()
```
